```python
def getNumeric_cols(df):
    """
    Input : DataFrame 
    Return Numeric Cols
    """
    ans = list()
    for col in df.columns:
        if (df[col].dtypes in [np.int64,np.float64]):
            ans.append(col)
    return ans
numericCols = getNumeric_cols(df_train)
```

```python
# Let's Visualize it Based on its Numeric cols first ,because its easy 
```

```python
#Lets Try to Focus to change data to ordinal or nominal ,lets do this
# i cant automate it yet , because the data is not really good to determine whether the current columns 
# is ordinal and nominal thus i make it manually
ordinalCols = ["ExterQual","ExterCond","BsmtQual","BsmtCond",
              "HeatingQC","KitchenQual","Functional","FireplaceQu","GarageQual",
               "GarageCond","PavedDrive","PoolQC"
              ]
note = ["PavedDrive","SaleCondition","MiscFeature","Fence","Electrical","BsmtExposure","BsmtFinType1",
        "BsmtFinType2"]
nominalCols = ["nanti ah Malirr"]
df_train[ordinalCols].head(1)
del df_train_PoolQCNonNullable
del df_train_PoolQCNullable
```

```python
df_train_copy = df_train.copy()
def convertOrdinalCols(val):
    val = str(val)
    if val.lower() == "ex":
        val = 5
    elif val.lower() == "gd":
        val = 4
    elif val.lower() == "ta":
        val =3
    elif val.lower() == "fa":
        val = 2
    elif val.lower() == "po":
        val = 1
    elif val.lower() == "nan":
        val = 0
    else:
        return np.nan
    return val
for i in ordinalCols:
    df_train_copy[i] = df_train_copy[i].apply(convertOrdinalCols)
df_train_copy.head(3)
```
