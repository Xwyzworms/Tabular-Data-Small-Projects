```python
from tqdm import tqdm
import pandas as pd
import numpy as np
```

```python
df = pd.read_json("contacts.json")
df.head()
```

```python
datNp = df.values
ans = {}
```

```python
def add_ans(user_id,value):
    if value != "":
        if value in ans:
            ans[value].add(user_id)
            
            return
        ans[value] = {user_id}
```

```python
ans
```

```python
for row in tqdm(datNp):
    user_id = row[0]
    add_ans(user_id, row[4])
    add_ans(user_id, row[1])
    add_ans(user_id, row[2])
```

```python
ans
```

```python
finalAns = {}
for ids in tqdm(ans.values() , total = len(ans.values())):
    currconnection = set(ids)
    
    for id in ids:
        if id in finalAns:
            currconnection.update(finalAns[id])
    
    for id in currconnection:
        finalAns[id] = currconnection
        
```
